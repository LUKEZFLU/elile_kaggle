## plant 1 generation

### Basic Cleaning

```python
# reading data plant 1 generation
plant1_g = pd.read_csv('./data/Plant_1_Generation_Data.csv')
print(plant1_g.shape)
plant1_g.sample(5)
```

```python
# Number of plant
len(plant1_g.PLANT_ID.unique())
```

```python
# Number of source key
len(plant1_g.SOURCE_KEY	.unique())
```

```python
# split time and date, then use YYYY_MM_DD
plant1_g[['DATE', 'TIME']] = plant1_g['DATE_TIME'].str.split(' ', expand=True)
plant1_g['DATE'] = pd.to_datetime(plant1_g['DATE'], format='%d-%m-%Y').dt.strftime('%Y-%m-%d')
plant1_g.sample(5)
```

### 2 single inverter (Use source key wCURE6d3bPkepu2 as an example)

```python
# key wCURE6d3bPkepu2
source_key = 'wCURE6d3bPkepu2'
sample = plant1_g[plant1_g['SOURCE_KEY'] == source_key]
sample_sorted = sample.sort_values(by=['DATE', 'TIME'])

sample_sorted


```

```python
# For a specific date
specific_date = '2020-05-15'
data_on_specific_date = sample_sorted[sample_sorted['DATE'] == specific_date]
data_on_specific_date
```
